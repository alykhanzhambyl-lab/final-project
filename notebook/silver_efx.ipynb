```python
from pyspark.sql import functions as F
fx0 = spark.table("dbo.bronze_ecb_fx_usd_eur_v2")

fx = fx0.select(F.to_date(F.col("TIME_PERIOD")).alias("day"), F.col("OBS_VALUE").cast("double").alias("usd_eur_rate")).dropna(subset=["day","usd_eur_rate"]).dropDuplicates(["day"])

fx.write.format("delta").mode("overwrite").option("overwriteSchema","true").saveAsTable("dbo.silver_ecb_fx_usd_eur_daily_v2")
print("fx rows:", spark.table("dbo.silver_ecb_fx_usd_eur_daily_v2").count())
spark.table("dbo.silver_ecb_fx_usd_eur_daily_v2").orderBy(F.col("day").desc()).show(5, truncate=False)
gdp0 = spark.table("dbo.bronze_worldbank_gdp_v2")

gdp = gdp0.select(F.col("date").cast("int").alias("year"), F.col("value").cast("double").alias("gdp_usd")).dropna(subset=["year","gdp_usd"])
gdp.write.format("delta").mode("overwrite").option("overwriteSchema","true").saveAsTable("dbo.silver_worldbank_gdp_us_v2")
print("gdp rows:", spark.table("dbo.silver_worldbank_gdp_us_v2").count())
spark.table("dbo.silver_worldbank_gdp_us_v2").orderBy(F.col("year").desc()).show(5, truncate=False)
```
